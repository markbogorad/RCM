# Merging Accounts and Contacts for Both Sheets
- matched acounts with matched contacts
- unmatched accounts with unmatched contacts

```python
matched_accounts["Provided Account ID"] = matched_accounts["Provided Account ID"].str.strip().str.upper()
matched_contacts["Provided Salesforce Contact Account Record ID"] = matched_contacts["Provided Salesforce Contact Account Record ID"].str.strip().str.upper()
print(matched_contacts["Provided Salesforce Contact Account Record ID"].isnull().sum())
print(matched_accounts["Provided Account ID"].isnull().sum())
```

```python
# Merge the DataFrames on the specified columns
merged_df = pd.merge(
    matched_accounts,
    matched_contacts,
    left_on="Provided Account ID",
    right_on="Provided Salesforce Contact Account Record ID",
    how="inner"
)

# Display the merged DataFrame
merged_df.to_excel("../data/prospect/matched.xlsx", index=False)

# Calculate and print match percentage
match_percentage = (len(merged_df) / len(matched_contacts)) * 100
print(f"Match rate: {match_percentage:.2f}%")
```

```python
unmatched_df = pd.merge(
    unmatched_accounts,
    unmatched_contacts,
    left_on="Account ID (Case Safe)",
    right_on="Account ID (Case Safe)",
    how="inner"
)

unmatched_df.to_excel("../data/prospect/unmatched.xlsx", index=False)

# Calculate and print match percentage
match_percentage = (len(unmatched_df) / len(unmatched_contacts)) * 100
print(f"Match rate: {match_percentage:.2f}%")
```

```python
unmatched_df = pd.merge(
    unmatched_accounts,
    unmatched_contacts,
    left_on="Account ID (Case Safe)",
    right_on="Account ID (Case Safe)",
    how="inner"
)

unmatched_df.to_excel("../data/prospect/unmatched.xlsx", index=False)

# Calculate and print match percentage
match_percentage = (len(unmatched_df) / len(unmatched_contacts)) * 100
print(f"Match rate: {match_percentage:.2f}%")
```
